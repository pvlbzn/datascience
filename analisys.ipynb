### Indexing and Extracting

```python
# To extract particular column
df['Churn']
```

```python

# Therefore we can take mean of it
df['Churn'].mean()

# 14.5% isnt good btw
```

```python

# Logical indexing
# df[P(df[<name>])]
# where P is some predicate
df[df['Churn'] == True].mean()

# Such that we can see the answer for the following question
# What is the average values of properties among non-loyal
# clients?
```

```python

# How long, during the day, non-loyal clients are using
# the phone?
df[df['Churn'] == True]['Total day minutes'].mean()
```

```python

# What is the minimal duration of international calls
# among loyal clients who doesn't use 'International plan'
# service?
df[(df['Churn'] == 0) & (df['International plan'] == 'No')]['Total intl minutes'].max()
```

```python

day_int_on  = df[
    df['International plan'] == 'Yes']['Total day charge'].mean()
day_int_off = df[
    df['International plan'] == 'No']['Total day charge'].mean()

# Clients with international plans spent more on day calls too
'$' + str(day_int_on - day_int_off)
```

```python

# Number of calls to costomer service performed by loyal
# clients, in average
df[df['Churn'] == False]['Customer service calls'].mean()
```

```python

# Non-loyal clients
df[df['Churn'] == True]['Customer service calls'].mean()
```

### Grouping
It has a general form of 
`df.grouoby(by=grouping_columns [columns_to_show].function()`

```python

# Group by Churn and show three cols
cols = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
df.groupby(['Churn'])[cols].describe(percentiles=[])
```

```python

# Or the same using agg method
df.groupby(['Churn'])[cols].agg([np.mean, np.std, np.min, np.max])
```
