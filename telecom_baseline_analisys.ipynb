### DataFrame Modifications
---

```python
# Count total number of calls for all users
total_calls = df['Total day calls'] + df['Total eve calls'] + df['Total night calls'] + df['Total intl calls']
df.insert(loc=len(df.columns), column='Total calls', value=total_calls)
df.head()
```

```python
# Or make it simpler
df['Total charge'] = df['Total day charge'] + df['Total eve charge'] + df['Total night charge'] + df['Total intl charge']

df.head()
```

```python
# Delete column
df = df.drop(['Total charge', 'Total calls'], axis=1)
df.head()
```

```python
# Delete rows
df = df.drop([1,2])
df.head()
```

### Churn Perdiction
---


```python
# Plot right in the notebook
%pylab inline

# Lets take a look how churn rate corellated to International Plan
churn_to_iplan = pd.crosstab(df['Churn'], df['International plan'], margins=True)
churn_to_iplan
```

```python
# Docs: http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html
churn_to_iplan.plot(kind='bar')
```

From this bar chart we can see that when International Plan is on the fraction of churn in higher.

```python
# Remove 'All' from dataset
churn_to_iplan = churn_to_iplan.drop(['All'], axis=1)
```

```python
churn_to_iplan.plot(kind='bar')
```

```python
# Number for service center tickets
c_to_service = pd.crosstab(df['Customer service calls'], df['Churn'], margins=True)
c_to_service = c_to_service.drop(['All'])
c_to_service
```

```python
c_to_service.plot(kind='bar')
```

Here we can clearly see that after 4th call to the support service we are likely to lost a client.

Lets elaborate on it:

```python
df['>3 Service Calls'] = (df['Customer service calls'] > 3).astype('int')
service_calls = pd.crosstab(df['>3 Service Calls'], df['Churn'], margins=True)
service_calls
```

```python
scd = service_calls.drop(['All'])
scd.plot(kind='bar')
```
